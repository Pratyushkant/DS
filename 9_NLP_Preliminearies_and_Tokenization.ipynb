# Natural Language Processing

Natural Language Processing (NLP) is a rapidly evolving field at the intersection of artificial intelligence, computer science, and linguistics. It focuses on enabling computers to understand, interpret, and generate human language in a valuable way. As AI continues to advance, NLP plays an increasingly crucial role in bridging the gap between human communication and machine understanding.

## Modern NLP: Goals and Fundamental Tasks

### The Modern Approach to NLP

Modern Natural Language Processing (NLP) has evolved significantly from its early days. The primary goal of modern NLP is not merely to understand language in a human-like manner, but rather to process and analyze language data to extract useful information and insights. This shift in focus has led to more practical and immediately applicable NLP systems.

#### Key Characteristics of Modern NLP:
1. **Data-Driven Approach**: Modern NLP relies heavily on large datasets and machine learning techniques.
2. **Task-Oriented**: Instead of aiming for general language understanding, modern NLP focuses on solving specific tasks effectively.
3. **Quantitative Output**: The goal is often to produce measurable, actionable insights from language input.

### Fundamental Tasks in NLP

NLP encompasses a wide range of tasks that are frequently encountered in various applications. These tasks can be broadly categorized into two main areas: Natural Language Understanding (NLU) and Natural Language Generation (NLG).

#### Natural Language Understanding (NLU)

NLU focuses on comprehending and analyzing input text. Key tasks include:

1. **Topic Modeling**
   - Question: "What is the topic of this text?"
   - Purpose: Identify the main themes or subjects discussed in a piece of text.
   - Applications: Content categorization, document clustering, trend analysis.

2. **Content Filtering**
   - Question: "Is this text inappropriate?"
   - Purpose: Detect and flag content that may be offensive, inappropriate, or violate certain guidelines.
   - Applications: Content moderation for social media, child-safe web browsing, email spam filtering.

3. **Sentiment Analysis**
   - Question: "Is this text positive, neutral, or negative?"
   - Purpose: Determine the emotional tone or attitude expressed in a piece of text.
   - Applications: Brand monitoring, customer feedback analysis, social media sentiment tracking.

4. **Named Entity Recognition (NER)**
   - Purpose: Identify and classify named entities (e.g., person names, organizations, locations) in text.
   - Applications: Information extraction, question answering systems, content indexing.

5. **Part of Speech (POS) Tagging**
   - Purpose: Assign grammatical categories (e.g., noun, verb, adjective) to each word in a text.
   - Applications: Text parsing, information extraction, machine translation.

6. **Information Retrieval**
   - Purpose: Find relevant information from a large collection of text data based on keywords or queries.
   - Applications: Search engines, document recommendation systems, knowledge management.

#### Natural Language Generation (NLG)

NLG involves creating human-readable text based on input data or prompts. Key tasks include:

1. **Language Modeling and Sentence Completion**
   - Question: "What is the next word or character?"
   - Purpose: Predict the most likely next word or sequence of words in a given context.
   - Applications: Autocomplete systems, predictive text, AI writing assistants.

2. **Machine Translation**
   - Question: "What is 'AI' in Tamil?"
   - Purpose: Translate text from one language to another while preserving meaning and context.
   - Applications: Cross-language communication, localization of content, global business operations.

3. **Text Summarization**
   - Question: "What is the crux of this paragraph?"
   - Purpose: Generate concise summaries of longer texts while retaining key information.
   - Applications: News briefings, document summarization, research paper abstracts.

4. **Question Answering**
   - Example: "Where is the nearest hair salon?"
   - Purpose: Generate relevant and accurate answers to specific questions based on available information.
   - Applications: Virtual assistants, customer support chatbots, information retrieval systems.

Modern NLP has shifted towards a more practical, task-oriented approach that focuses on extracting useful information and generating valuable insights from language data. By breaking down language processing into these fundamental tasks, NLP systems can be more effectively designed and implemented to solve real-world problems across various domains, from business intelligence to personal digital assistants.

NLU can be solved as a subproblem of NLG using extractive summary.

- Extractive summarization, typically considered an NLU task, involves selecting the most important sentences or phrases from a source text to create a summary.
- By framing NLG tasks as extractive problems, we can leverage NLU techniques to generate text by selecting and combining existing content rather than creating entirely new text.
- This approach can be particularly effective for tasks like question answering or text summarization, where the desired output often exists within the input text or a knowledge base.
- For example, in question answering, the system can use NLU to understand the question and then extract the most relevant sentence or passage from a corpus as the answer.
- While this method may not be suitable for all NLG tasks, especially those requiring novel content generation, it can be an efficient and effective approach for many practical applications, bridging the gap between understanding and generation in NLP.

### Challenges in NLP

Despite significant progress, NLP faces several challenges:

1. **Ambiguity in Language**: Words and phrases can have multiple meanings depending on context, making it difficult for machines to interpret accurately.

2. **Contextual Understanding**: Grasping the broader context and subtext in communication remains a significant challenge for AI systems.

3. **Handling Colloquialisms and Idioms**: Informal language, slang, and culturally-specific expressions are often difficult for NLP systems to interpret correctly.

4. **Multilinguality**: Developing systems that can effectively process and understand multiple languages is complex.

5. **Maintaining Conversation Flow**: For chatbots and AI assistants, maintaining coherent, context-aware conversations over extended interactions is challenging.

6. **Bias in Language Models**: NLP models can inadvertently learn and perpetuate biases present in their training data.

### Recent Advancements

1. **Transformer Models**: The introduction of transformer architecture, as seen in models like BERT and GPT, has revolutionized NLP tasks by enabling better understanding of context and long-range dependencies.

2. **Few-Shot Learning**: Advanced language models can now perform tasks with minimal task-specific training, greatly enhancing their versatility.

3. **Multimodal NLP**: Integrating text with other data types like images and audio for more comprehensive language understanding.

4. **Explainable AI in NLP**: Developing techniques to make NLP model decisions more interpretable and transparent.

5. **Efficient NLP**: Creating smaller, more efficient models that can run on edge devices without compromising significantly on performance.

### Naive Bayes for Spam Classification and LDA Adaptation

#### Naive Bayes for Spam Classification

For instance,

1. **Data:**
   - Spam emails: 10,000
   - Not spam emails: 90,000
   - Dictionary size: 20,000 words
   - New email length: 25 words

2. **Probability Calculation:**
   The probability of an email being spam given its words is proportional to:

   $$
   P(spam | word1, word2, ..., word25) ∝ P(spam) * P(word1|spam) * P(word2|spam) * ... * P(word25|spam)
   $$

   This is based on Bayes' theorem and the "naive" assumption that words are independent given the class (spam or not spam).

3. **Implementation:**
   - Calculate P(spam) and P(not spam) from the dataset.
   - For each word in the vocabulary, calculate P(word|spam) and P(word|not spam).
   - For a new email, multiply these probabilities for each word in the email.
   - Compare the final probability for spam vs. not spam to classify the email.

#### Adapting to Latent Dirichlet Allocation (LDA)

While Naive Bayes is a classification algorithm and LDA is a topic modeling technique, we can draw some parallels in the approach:

1. **Document-Topic Distribution:**
   Instead of P(spam) and P(not spam), LDA would have P(topic_k) for each topic k.

2. **Word-Topic Distribution:**
   Instead of P(word|spam), LDA would have P(word|topic_k) for each word and topic.

3. **Generative Process:**
   - For each document:
     - Choose a topic distribution (analogous to choosing spam or not spam)
     - For each word:
       - Choose a topic from the document's topic distribution
       - Choose a word from that topic's word distribution

4. **Inference:**
   - Use techniques like Gibbs sampling or variational inference to estimate:
     - The topic distribution for each document
     - The word distribution for each topic

5. **Key Differences:**
   - LDA allows for multiple topics per document, unlike the binary classification in spam detection.
   - LDA is unsupervised, discovering topics without labeled data, whereas Naive Bayes for spam detection is supervised.
   - LDA uses more complex probabilistic models and inference techniques compared to the simpler Naive Bayes approach.

By understanding the probabilistic nature of Naive Bayes, one can begin to grasp the more complex probabilistic model used in LDA, where instead of just two classes (spam/not spam), we have multiple topics, and documents can be mixtures of these topics.

## Tokenization in Natural Language Processing

Tokenization is a fundamental task in Natural Language Processing (NLP) that involves breaking down text into smaller units called tokens. These tokens are typically words, but can also be characters, subwords, or phrases depending on the specific application.

### Why Tokenization is Important

1. **Preprocessing**: It's often the first step in text processing pipelines.
2. **Feature extraction**: Tokens serve as the basic units for further analysis.
3. **Standardization**: It helps in standardizing text for consistent processing.
4. **Meaning preservation**: Proper tokenization maintains the semantic meaning of the text.

Here's the completed list of types of tokenization, including the "Provided Vocabulary Based" tokenization:

### Types of Tokenization

1. **Word Tokenization**: Splits text into individual words.
2. **Sentence Tokenization**: Divides text into sentences.
3. **Subword Tokenization**: Breaks words into meaningful subunits (useful for handling unknown words).
4. **Character Tokenization**: Splits text into individual characters (used in some deep learning models).
5. **Provided Vocabulary Based**: Tokenizes text based on a pre-defined vocabulary, often used in machine learning models. This approach maps words or subwords to unique IDs from a fixed vocabulary. It typically handles out-of-vocabulary words using special tokens or subword segmentation.

Additional types of tokenization include:

6. **Whitespace Tokenization**: Splits text based on whitespace characters. It's simple but can be ineffective for languages without clear word boundaries.

7. **N-gram Tokenization**: Creates tokens of N contiguous words or characters, useful for capturing local context.

8. **Regular Expression Based Tokenization**: Uses custom regular expressions to define token boundaries, allowing for flexible, rule-based tokenization.

9. **Byte-Pair Encoding (BPE)**: A subword tokenization method that iteratively merges the most frequent pair of bytes or characters to build a vocabulary of subword units.

10. **WordPiece Tokenization**: Similar to BPE, but uses a likelihood criterion to select merge operations. It's commonly used in models like BERT.

11. **SentencePiece Tokenization**: Treats the input as a raw stream of characters and learns to tokenize in an unsupervised manner, useful for languages without clear word boundaries.

12. **Morphological Tokenization**: Splits words into morphemes (the smallest units of meaning), which can be beneficial for highly inflected languages.

Each type of tokenization has its strengths and is suited for different languages, tasks, or model architectures in natural language processing.

## Byte Pair Encoding in NLP

In the rapidly evolving field of Natural Language Processing (NLP), subword embedding techniques have emerged as a game-changer. Among these, Byte Pair Encoding (BPE) stands out as a simple yet powerful method that has become a cornerstone of modern deep learning-based NLP models. Let's dive deep into what BPE is, how it works, and why it's so important.

### What is Byte Pair Encoding?

Byte Pair Encoding is a data compression technique that was originally developed in the 1990s. However, its application to NLP, particularly for subword tokenization, has revolutionized the way we process and understand text data.

### The Power of BPE in Modern NLP

BPE powers many modern deep learning-based NLP models. Its strength lies in its ability to balance the trade-off between having a manageably sized vocabulary and capturing meaningful subword units. This is particularly useful for:

1. Handling out-of-vocabulary words
2. Capturing morphological information
3. Dealing with multiple languages efficiently

### How BPE Works: A Simple Yet Powerful Idea

The core idea behind BPE is remarkably straightforward:

1. Start with individual characters (or bytes) as the initial set of tokens.
2. Identify the most frequently occurring pair of adjacent tokens.
3. Merge this pair to create a new token.
4. Repeat steps 2 and 3 until you reach a desired vocabulary size.

While the concept is simple, its implications for NLP are profound.

#### The BPE Algorithm in Detail

1. **Initialization**: Begin with UTF-8 bytes as tokens (256 in total).
2. **Frequency Analysis**: Identify the most frequently occurring byte-pair in the dataset.
3. **Merge Operation**: Add this byte-pair as a new token to the vocabulary.
4. **Iteration**: Keep merging until you hit the desired vocabulary size (a hyperparameter).

#### An Illustrative Example

Let's walk through a simplified example to see BPE in action:

Initial data: aaabdaaabac

1. 'aa' occurs most frequently, so we replace it with 'Z':
   Result: ZabdZabac

2. 'ab' is now the most frequent pair, so we replace it with 'Y':
   Result: ZYdZYac

3. 'ZY' becomes the most frequent pair, so we replace it with 'X':
   Result: XdXac

In this example, we've reduced our original 11-character string to a 5-character representation using just three merge operations.

#### Why BPE Matters

1. **Vocabulary Size Control**: BPE allows us to precisely control the size of our vocabulary, balancing between model complexity and expressive power.

2. **Handling Rare Words**: By breaking words into subword units, BPE can handle rare or unseen words effectively, improving model generalization.

3. **Language Agnostic**: BPE works well across different languages, including those without clear word boundaries.

4. **Morphological Awareness**: It can capture meaningful morphemes, helping models understand word structure.

#### Challenges and Considerations

While BPE is powerful, it's not without challenges:

1. **Hyperparameter Sensitivity**: The choice of vocabulary size can significantly impact model performance.

2. **Loss of Interpretability**: Subword tokens can sometimes be less interpretable than full words.

3. **Computational Overhead**: The tokenization process adds some computational cost during both training and inference.

#### Conclusion

Byte Pair Encoding has become an integral part of the NLP toolkit, driving significant improvements in various tasks from machine translation to text classification. Its simplicity, coupled with its effectiveness, makes it a fascinating example of how fundamental ideas can have profound impacts on advanced technologies.

As NLP continues to evolve, techniques like BPE remind us that sometimes, the most powerful solutions are built on elegantly simple foundations. Understanding and leveraging these methods is key to pushing the boundaries of what's possible in natural language understanding and generation.

# Word-Level Tokenization and One-Hot Encoding in Natural Language Processing

## 1. Word-Level Tokenization

Word-level tokenization is a fundamental process in Natural Language Processing (NLP) that involves breaking down a given text into individual words or tokens. This process is crucial for converting human-readable text into a format that can be processed by machine learning algorithms.

### 1.1 Vocabulary Construction

The first step in word-level tokenization is constructing a vocabulary:

- **Definition**: A vocabulary is a comprehensive list of all unique words that appear in the corpus or are expected to appear in future texts.
- **Size**: For the English language, a vocabulary of approximately 20,000 words typically covers most commonly used terms.
- **Special Tokens**: The vocabulary often includes special tokens such as:
  - `<UNK>`: Represents unknown words not present in the vocabulary.
  - `<PAD>`: Used for padding sequences to a fixed length.
  - `<START>` and `<END>`: Markers for the beginning and end of sequences.

### 1.2 Tokenization Process

Once the vocabulary is established, the tokenization process involves:

1. Splitting the input text into individual words.
2. Matching each word to its corresponding index in the vocabulary.
3. Representing the text as a sequence of these indices.

Example:
```
Input: "I love NLP"
Tokenized: [20, 212, 559]
```
Where 20, 212, and 559 are the hypothetical indices of "I", "love", and "NLP" in the vocabulary, respectively.

## 2. One-Hot Encoding

One-hot encoding is a method of representing categorical data as binary vectors.

### 2.1 Definition

A one-hot vector is a binary vector with a length equal to the size of the vocabulary, where:
- The index corresponding to the word is set to 1.
- All other indices are set to 0.

### 2.2 Process

For each token index obtained from the tokenization process:
1. Create a vector of zeros with length equal to the vocabulary size.
2. Set the element at the index corresponding to the token to 1.

Example:
For a vocabulary of size 20,000:
```
"I" (index 20) → [0, 0, ..., 1, ..., 0]
                             ^
                     20th position
```

### 2.3 Characteristics

- **Dimensionality**: Each vector has a dimension equal to the vocabulary size.
- **Sparsity**: One-hot vectors are sparse, with only one non-zero element.
- **Orthogonality**: All one-hot vectors are orthogonal to each other.

## 3. From Tokenization to Vector Representation

The complete process from raw text to vector representation involves:

1. Tokenizing the input text using the predefined vocabulary.
2. Converting each token to its corresponding index.
3. Transforming each index into a one-hot vector.

This results in a sequence of one-hot vectors, each representing a word in the original text.

## 4. Limitations and Alternatives

While one-hot encoding is simple and widely used, it has limitations:
- High dimensionality, especially for large vocabularies.
- Lack of semantic information about words.

Alternatives include:
- Word embeddings (e.g., Word2Vec, GloVe)
- Contextual embeddings (e.g., BERT, ELMo)

These methods provide dense vector representations that capture semantic relationships between words, addressing some of the limitations of one-hot encoding.

# Document Representation with Fixed-Length Context and One-Hot Encoding

## 1. Fixed-Length Context

In Natural Language Processing (NLP), documents are often represented using a fixed-length context to standardize input size for machine learning models.

### 1.1 Context Length

- **Definition**: The maximum number of tokens considered for processing a document.
- **Typical values**: 512, 1000, or 1024 tokens.
- **Purpose**: To create a uniform input size for NLP models, regardless of the original document length.

### 1.2 Handling Document Length

1. **Documents shorter than context length**:
   - Filled with padding tokens (usually zeros) to reach the fixed length.
   - Example: If context length is 512 and document has 100 tokens, 412 padding tokens are added.

2. **Documents longer than context length**:
   - Truncated to fit the context length.
   - Two common approaches:
     a. First-In-First-Out (FIFO): Keep the first N tokens, where N is the context length.
     b. Last-In-First-Out (LIFO): Keep the last N tokens.

## 2. Matrix Representation

Documents are represented as matrices to facilitate processing by machine learning models.

### 2.1 Matrix Dimensions

- Rows: Vocabulary size
- Columns: Maximum document length (context length)

### 2.2 One-Hot Encoding in Matrix Form

Each column in the matrix represents a token in the document:
- The row corresponding to the token's vocabulary index is set to 1.
- All other rows are set to 0.

### 2.3 Example

For a vocabulary of size 20,000 and a context length of 512:
- Matrix dimensions: 20,000 × 512
- For the sentence "I love NLP":
  - Column 1: One-hot vector for "I"
  - Column 2: One-hot vector for "love"
  - Column 3: One-hot vector for "NLP"
  - Columns 4-512: All zeros (padding)

## 3. Limitations of One-Hot Encoding for Word Representation

One-hot encoding, while simple and widely used, has several significant limitations when used for word representation in Natural Language Processing (NLP). This document outlines the key issues and their implications.

### 3.1. Sparsity and High Dimensionality

#### 3.1.1 Problem Description

- **Sparsity**: One-hot vectors are extremely sparse, with only one non-zero element.
- **High Dimensionality**: The dimension of each vector is equal to the vocabulary size, which can be very large (e.g., 20,000 or more).

#### 3.1.2 Visual Representation

If visualized as a matrix:
- Rows: Words in the vocabulary
- Columns: Positions in the document
- Content: Mostly zeros (black) with occasional ones (white)

#### 3.1.3 Implications

- Inefficient use of memory and computational resources
- Difficulty in capturing semantic relationships between words
- Challenges in scaling to large vocabularies or datasets

### 3.2. Hard-Coded Nature

#### 3.2.1 Problem Description

- The position of the '1' in each vector is predetermined by the word's index in the vocabulary.
- This representation is fixed and does not adapt to the context or usage of words.

#### 3.2.2 Implications

- Lack of flexibility in representing words with multiple meanings
- Inability to capture evolving language use or domain-specific terminology
- Difficulty in handling out-of-vocabulary words

### 3.3. Orthogonality and Independence Assumption

#### 3.3.1 Mathematical Property

- Any two different one-hot vectors are orthogonal to each other.
- Dot product between any two different one-hot vectors is always zero.

#### 3.3.2 Linguistic Implication

- This property mathematically implies that all words are independent of each other.

#### 3.3.3 Problem with this Assumption

- In natural language, words are not independent:
  - Synonyms (e.g., "happy" and "joyful") are closely related
  - Antonyms (e.g., "hot" and "cold") have a strong relationship
  - Words in the same semantic field (e.g., "dog", "cat", "pet") are related
- This representation fails to capture these semantic relationships

### Example

Consider the words "king", "queen", "man", and "woman":
- In one-hot encoding, these would be represented as four orthogonal vectors.
- However, there are clear semantic relationships between these words (royalty, gender) that are not captured.

### 3.4. Lack of Contextual Information

- One-hot encoding represents each word in isolation, without considering its context in a sentence or document.
- This leads to an inability to capture:
  - Word sense disambiguation
  - Contextual nuances
  - Phrasal meanings

### 3.5. Implications for NLP Tasks

These limitations affect various NLP tasks:
- **Semantic Analysis**: Difficulty in measuring word similarity or relatedness
- **Machine Translation**: Challenges in capturing nuanced meanings across languages
- **Text Classification**: Potential loss of important semantic information

### 3.6. Alternative Approaches

To address these limitations, several alternative approaches have been developed:
1. **Word Embeddings** (e.g., Word2Vec, GloVe): Dense vector representations that capture semantic relationships
2. **Contextual Embeddings** (e.g., BERT, ELMo): Dynamic representations that change based on the word's context
3. **Subword Tokenization**: Methods like BPE or WordPiece that can handle out-of-vocabulary words and capture morphological information

These approaches aim to create more nuanced, context-aware, and semantically rich representations of words, addressing many of the limitations inherent in one-hot encoding.

# Word Embeddings and Geometric Relationships in Semantic Space

## 1. Desired Properties of Word Representations

### 1.1 Geometric Relationship Reflecting Semantic Relationship

The primary goal of modern word representation techniques is to create a mapping where the geometric relationships between word vectors in a high-dimensional space correspond to the semantic relationships between the words they represent.

### 1.2 L2 Distance and Cosine Distance

Two primary metrics are used to measure the relationship between word vectors:

- **L2 Distance (Euclidean Distance)**: Measures the straight-line distance between two points in the vector space.
- **Cosine Distance**: Measures the angle between two vectors, indicating their directional similarity.

The key principle is that these geometric distances should correlate with the "semantic distance" between words. For example:
- Similar words (e.g., "cat" and "kitten") should have vectors that are close together.
- Dissimilar words (e.g., "cat" and "astronomy") should have vectors that are far apart.

## 2. Word Embeddings

### 2.1 Definition

Word embeddings are dense vector representations of words that map human language into a structured geometric space.

### 2.2 Key Characteristics

- **Dimensionality**: Typically range from 50 to 300 dimensions, much lower than one-hot encodings.
- **Density**: Unlike sparse one-hot vectors, every dimension typically has a non-zero value.
- **Learned**: These representations are learned from large corpora of text, capturing semantic and syntactic relationships.

### 2.3 Geometric Properties

In the embedding space:
- Words with similar meanings cluster together.
- Semantic relationships are often reflected in vector arithmetic. For example:
  - vec("king") - vec("man") + vec("woman") ≈ vec("queen")

### 2.4 Advantages Over One-Hot Encoding

1. **Captures Semantics**: Similar words have similar vectors.
2. **Efficient**: Lower dimensionality makes computations more efficient.
3. **Generalization**: Allows models to generalize better to unseen words or rare words.

## 3. Examples of Semantic Relationships in Embedding Space

### 3.1 Synonyms and Antonyms

- Synonyms (e.g., "happy" and "joyful") will have vectors close to each other.
- Antonyms (e.g., "hot" and "cold") may be close in some dimensions but differ significantly in others.

### 3.2 Analogies

Word embeddings can capture analogical relationships. For example:
- "man" is to "woman" as "king" is to "queen"
- This relationship is reflected in the vector space, allowing for arithmetic operations on word vectors.

### 3.3 Semantic Fields

Words belonging to the same semantic field (e.g., colors, animals, or professions) tend to cluster together in the embedding space.

## 4. Popular Word Embedding Techniques

1. **Word2Vec**: Uses shallow neural networks to learn word associations from a large corpus of text.
2. **GloVe (Global Vectors)**: Combines global matrix factorization and local context window methods.
3. **FastText**: Extends Word2Vec by using subword information, allowing for better representation of rare words and out-of-vocabulary words.

## Sequence model approach

The sequence model approach involves representing input text as a sequence of integers (one per word), mapping these integers into a vector space using embeddings, and then feeding the resulting sequence of vectors into a stack of layers capable of handling sequences, such as Transformers or recurrent neural networks (RNNs), including LSTMs and bidirectional LSTMs (Bi-LSTMs).

## Types of word embeddings

Word embeddings can be categorized into two types: contextual and non-contextual.

- Contextual embeddings (e.g., ELMo and BERT) vary based on the word's location in a sentence, allowing for dynamic representations.
- Non-contextual embeddings (e.g., Word2Vec and GloVe) provide the same vector representation for a word, regardless of its context in a sentence.

Both ELMo and BERT utilize transformer architecture for training, whereas Word2Vec and GloVe generate fixed vectors for words.

# Language Model

- Any model that can predict the probability of the next token in a sequence of text input (converted to embeddings) is called a Language Model  
  $$*p(\text{next token} | \text{all previous tokens})*$$

- LM captures the latent space of language: its statistical structure

- Large Language Models are trained on large text corpora (trillions of tokens) and have billions of parameters

- They have emergent abilities:
  - Can do tasks for which it is not explicitly trained
  - Today, we don’t take a chance and make it learn to follow instructions, called as instruction training


## Masked Language Modeling (MLM)

- Random tokens in a document are masked.
- Neural models predict the masked tokens based on context.
- MLM is a fill-in-the-blank task.

**Example:**
  - Original: *The cat sits on the mat.*  
  - Masked: *The [MASK] sits on the mat.*  
  - Task: Predict "cat" given the surrounding words.

- Objective: Maximize \( p(\text{masked token} \mid \text{all other tokens}) \).

---

## Sentence Completion Modeling (Next Token Prediction)

- Uses an **autoregressive** setup.
- At each inference step, the model predicts the next token from the vocabulary as a probability distribution.
  
**Prediction Process:**
  - \((k+1)^{st}\) token: Input = (prompt + predicted k tokens)  
  - \((k+2)^{nd}\) token: Input = (prompt + predicted k+1 tokens)  

## Pretraining

- **Pre-training** initializes weights and biases by training models on large datasets using tasks like **Masked Language Modeling (MLM)** or **Next Token Prediction**.
- In **MLM**, random tokens are masked (e.g., "IISc is located in [MASK] India"), and the model fills the blank (e.g., "Bangalore"), learning trivia, syntax, and coreferences.
- **Next Token Prediction** predicts the next word in a sequence, functioning as a "fill-in-the-blank" task at the sentence's end.
- Both MLM and Next Token Prediction help models understand **semantic structure**, **syntax**, and **sentiment**.
- The models can also handle **coreference resolution** (e.g., "Naruto is the Hokage... he feels responsible"), **topic modeling**, and **reasoning** from implicit information.
- Pre-training is analogous to image processing, where models trained from scratch on datasets like **ImageNet** are later fine-tuned through **transfer learning**.
- Fine-tuning involves adapting the pre-trained model to perform specific tasks like sentiment analysis or token generation more effectively.
- The neural net models that are capable of doing fill in the blanks are called as transformer encoder models and the neural models that are capable of doing next token prediction are called as transformer decoder class of models

````python
import os
import re
import requests
from collections import Counter
from sklearn.decomposition import PCA

import numpy as np 
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
````

````python
# Tokenization and vocabulary building
def tokenize(text):
    return re.findall(r'\w+', text.lower()) # Convert to lowercase

def build_vocab_test(dataset, max_size=10000):
    """Builds a vocabulary from the dataset."""
    word_counts = Counter() # Counts the frequency of each word
    for text in dataset:
        word_counts.update(tokenize(text)) # Update the counter with the words in the text
    
    # Create vocabulary with '<unk>' and most common words
    return ["<unk>"] + [word for word, _ in word_counts.most_common(max_size - 1)] # -1 to make space for '<unk>'

# Example dataset
dataset = [
    "I write, erase, rewrite",
    "Erase again, and then",
    "A poppy blooms.",
]

# Build the vocabulary
vocabulary = build_vocab_test(dataset) # Build vocabulary
print(f"vocabulary = {vocabulary}") # Print vocabulary
print(f"len(vocabulary) = {len(vocabulary)}") # Print vocabulary length
print(f"Is 'still' in the vocabulary? {'still' in vocabulary}")
test_sentence = "I write, rewrite, and still rewrite again"
print(f"Is 'still' in the test_sentence? {'still' in tokenize(test_sentence)}")
print(f"len(tokenize(test_sentence)) = {len(tokenize(test_sentence))}")
````

````python
word_to_idx = {word: idx for idx, word in enumerate(vocabulary)} # Create a dictionary that maps words to indices
print(word_to_idx)
````

````python
def vectorize(test_sentence, vocab):
    """Converts a test sentence into a vector based on the vocabulary."""
    # Tokenize the test sentence
    tokens = tokenize(test_sentence)
    # Map each token to its corresponding index in the vocabulary
    vector = [vocab.index(token) if token in vocab else vocab.index("<unk>") for token in tokens] # Use the index of '<unk>' if the token is not in the vocabulary
    return vector

# Build the vocabulary (as defined previously)
vocabulary = build_vocab_test(dataset)

# Vectorize the test sentence
vectorized_sentence = vectorize(test_sentence, vocabulary)
print("Vectorized Sentence:", vectorized_sentence)
````

````python
def download_imdb():
    url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
    if not os.path.exists("aclImdb_v1.tar.gz"):
        print("Downloading IMDB dataset...")
        response = requests.get(url)
        with open("aclImdb_v1.tar.gz", "wb") as f:
            f.write(response.content)
        print("Download complete. Extracting...")
        import tarfile
        with tarfile.open("aclImdb_v1.tar.gz", "r:gz") as tar:
            tar.extractall()
        print("Extraction complete.")

download_imdb()
````

````python
def build_vocab(file_list, max_size=10000):
    word_freq = Counter()
    
    for file_path in file_list:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()
            # Tokenize the text (split into words)
            words = re.findall(r'\b\w+\b', text)
            word_freq.update(words)
    
    # Select the most common words up to max_size
    vocab = [word for word, _ in word_freq.most_common(max_size)]
    
    return vocab
````

````python
# Load IMDB dataset
def load_imdb(split='train'):
    data = []
    for label in ['pos', 'neg']:
        path = f'aclImdb/{split}/{label}'
        for file in os.listdir(path):
            with open(os.path.join(path, file), 'r', encoding='utf-8') as f:
                data.append((f.read(), 1 if label == 'pos' else 0))
    return data
````

````python
# Build vocabulary
train_files = [os.path.join('aclImdb/train/pos', f) for f in os.listdir('aclImdb/train/pos')] + \
              [os.path.join('aclImdb/train/neg', f) for f in os.listdir('aclImdb/train/neg')]
vocab = build_vocab(train_files)
word_to_idx = {word: idx for idx, word in enumerate(vocab)}
````

````python
print(word_to_idx)
````

````python
# Function to convert text to tensor
def text_to_tensor(text, max_length=600):
    tokens = tokenize(text)
    ids = [word_to_idx.get(token, 0) for token in tokens]
    if len(ids) < max_length:
        ids += [0] * (max_length - len(ids))
    else:
        ids = ids[:max_length]
    return torch.tensor(ids, dtype=torch.long)
````

````python
# Custom dataset class
class IMDBDataset(Dataset):
    def __init__(self, data):
        self.data = [(text_to_tensor(text), label) for text, label in data]
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return self.data[idx]
````

````python
# Create datasets and dataloaders
train_data = load_imdb('train')
test_data = load_imdb('test')

train_dataset = IMDBDataset(train_data)
test_dataset = IMDBDataset(test_data)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)
````

````python
import torch.nn.functional as F
# Define the model (same as before)
class SentimentModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim=256):
        super().__init__()

        # Q: What is the input to the Embedding layer?
        # The imput to the embedding layer is the one-hot encoded vector of the word
        # Q: What is the dimension of the output embeddings?
        # The dimension of the output embeddings is 256
        # Q: In embedding layer shape, what are None and None?
        # The first None is the batch size and the second None is the number of words in the sentence

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        # Add a Dense layer with 64 neurons
        self.fc1 = nn.Linear(embedding_dim, 64)
        # YOUR CODE HERE to add a Dense layer with 32 neurons
        self.fc2 = nn.Linear(64, 32)
        # YOUR CODE HERE to add a Dense layer with 16 neurons
        self.fc3 = nn.Linear(32, 16)
        # YOUR CODE HERE to add a Dense layer with 1 neuron
        self.fc4 = nn.Linear(16, 1)
        # YOUR CODE HERE to add a Dropout layer
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x):
        # Embedding layer
        x = self.embedding(x)
        
        # Average pooling over the sequence length
        x = torch.mean(x, dim=1)
        
        # Fully connected layers with ReLU activation and dropout
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = F.relu(self.fc3(x))
        x = self.dropout(x)
        x = self.fc4(x)
        
        return x.squeeze()
````

````python
model = SentimentModel(len(vocab)).to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.RMSprop(model.parameters())

print(model)
````

````python
# Function to visualize word embeddings (same as before)
def visualize_embeddings(model, vocab, words):
   
   ## YOUR CODE Calculate embedding vectors and do dimensional reduction to 2 Dimension using PCA
    embeddings = model.embedding.weight.data.cpu().numpy()
    reduced_embeddings = PCA(n_components=2).fit_transform(embeddings)

    plt.figure(figsize=(10, 8))
    for word in words:
        idx = word_to_idx.get(word, 0)
        x, y = reduced_embeddings[idx]
        plt.scatter(x, y)
        plt.annotate(word, (x, y))
    plt.title("Word Embeddings Visualization")
    plt.show()

test_words = ['good', 'bad', 'nice', 'poor', 'terrible', 'terrific', 'awesome', 'awful', 'best', 'worst']
visualize_embeddings(model, vocab, test_words)
````

````python
# Training function (same as before)
def train_model(model, train_loader, val_loader, epochs=20):
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs.squeeze(), labels.float())
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs.squeeze(), labels.float())
                val_loss += loss.item()
                predicted = (outputs.squeeze() > 0).int()
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        
        print(f"Epoch {epoch+1}/{epochs}")
        print(f"Train Loss: {train_loss/len(train_loader):.4f}")
        print(f"Val Loss: {val_loss/len(val_loader):.4f}")
        print(f"Val Accuracy: {correct/total:.4f}")

# Train the model
train_model(model, train_loader, test_loader)
````

````python
test_words = ['good', 'bad', 'nice', 'poor', 'terrible', 'terrific', 'awesome', 'awful', 'best', 'worst']
visualize_embeddings(model, vocab, test_words)
````

````python
# Evaluate the model (same as before)
model.eval()
test_loss = 0.0
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        loss = criterion(outputs.squeeze(), labels.float())
        test_loss += loss.item()
        predicted = (outputs.squeeze() > 0).int()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Test Loss: {test_loss/len(test_loader):.4f}")
print(f"Test Accuracy: {correct/total:.4f}")
````
